## Cleaning Columns

```python
autos.columns
```

```python
autos_copy = autos.copy()
```

```python
## Changing the names of the columns to add clarity##
autos_copy.rename({"yearOfRegistration" : "registration_year", "monthOfRegistration" : "registration_month", "notRepairedDamage" : "unrepaired_damage", "dateCreated" : "ad_created"}, axis = 1, inplace = True)
```

```python
## function from Geeks for Geeks that converts a string from Camelcase to Snakecase
def change_case(str): 
    res = [str[0].lower()] 
    for c in str[1:]: 
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'): 
            res.append('_') 
            res.append(c.lower()) 
        else: 
            res.append(c) 
      
    return ''.join(res) 

## loop over columns applying the above function and appending to copied dataframe
new_columns = []
for c in autos_copy.columns:
    clean_c = change_case(c)
    new_columns.append(clean_c)
    
autos_copy.columns = new_columns
```

```python
autos_copy.head()
```

Above I viewed the columns, noting that they were all in [Camel Case](https://en.wikipedia.org/wiki/Camel_case) and felt that some column names were confusing. 

I first cleaned up the confusing column names and renamed them in snakecase. For the rest of the column names, I found a function from [Geeks for Geeks](https://www.geeksforgeeks.org/python-program-to-convert-camel-case-string-to-snake-case/) as I was not sure how to build a function that would covert strings from Camel Case to Snake Case. ## Summary of Brands
There are 40 different car brands listed for the time period. German car brands (volkswagen, bms, opel, mercedes_benz, and audi) make up the top 5 most popular brands with volkswagen being the most popular brand. The number of volkswagens is nearly double the number of bmws, the second most popular brand. 

Becuase there are many brands that make up a small percentage of the listings, I will focus on brands with 1000+ listings (2.5 percent or more) to get an idea of the average prices per most popular brands. 

```python
brand_percentage = autos_copy["brand"].value_counts(normalize = True)
popular_brands = brand_percentage[brand_percentage > .025].index
print(popular_brands)
```

```python
brand_mean_price = {}

for brand in popular_brands:
    selected_brand = autos_copy[autos_copy["brand"] == brand]
    mean_price = selected_brand["price_dollars"].mean()
    brand_mean_price[brand] = int(mean_price)
    
brand_mean_price
```
